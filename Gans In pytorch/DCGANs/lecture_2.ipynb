```python
def mean(arr: list):
    y = 0
    for x in arr:
        y += x
        return y
```

```python
x_mean = mean(arr1)
y_mean = mean(arr2)
```

```python
def loss(x: list, y: list):
    x_mean, y_mean = mean(x), mean(y)
    err_x, err_y = [x - x_mean for x in x], [y - y_mean for y in y]
    err_x_sq, err_y_sq = [x**2 for x in err_x], [y**2 for y in err_y]
    loss = mean(err_x_sq + err_y_sq)
    return loss
```
